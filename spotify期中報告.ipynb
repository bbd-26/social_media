# 環境建置

```python
import os
import requests
import json
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pprint

acc_token = ""
```

## 整理json檔的函式

```python
def json_sorting(json_data):
    """
    將 JSON 資料轉換為 Python 可用的程式碼格式。

    :param json_data: 字串格式或字典格式的 JSON 資料
    :return: Python 程式碼字串
    """
    # 如果輸入是字串格式，將其轉換為字典
    if isinstance(json_data, str):
        try:
            json_data = json.loads(json_data)
        except json.JSONDecodeError:
            raise ValueError("無效的 JSON 字串")

    # 使用 pprint 來格式化字典
    python_code = pprint.pformat(json_data, indent=4)
    return python_code
```

## 簡化請求api的函式

```python
def request_api(access_token, rq_type, obj_id):
  # 請求的 URL 和 Access Token
  url = "https://api.spotify.com/v1/" + rq_type + "/" + obj_id
  access_token = access_token

  # 設置請求頭
  headers = {
      "Authorization": f"Bearer {access_token}"
  }

  # 發送 GET 請求
  response = requests.get(url, headers=headers)

  # 檢查回應
  if response.status_code == 200:
      # 打印 artist 資訊
      response_data = response.json()
      print(json_sorting(response_data))
  else:
      print("Request failed:", response.status_code, response.text)
```

# 產生access token

產生 access token 需要cilent id 和 secret，token有效時間為一小時

```python
client_id = "48540c92930b4ac2884f64f43de91be5"
client_secret = ""

# 請求的 URL 和數據
url = "https://accounts.spotify.com/api/token"
data = {
    "grant_type": "client_credentials",
    "client_id": client_id,
    "client_secret": client_secret
}

headers = {
    "Content-Type": "application/x-www-form-urlencoded"
}

# 發送 POST 請求
response = requests.post(url, headers=headers, data=data)

# 檢查回應
if response.status_code == 200:
    print("Access token:", response.json().get("access_token"))
    acc_token = response.json().get("access_token")
else:
    print("Failed to get token:", response.status_code, response.text)
```

# search相關API

* q=tag:hipster 搜尋受歡迎程度最低的 10% 的專輯
* q=tag:new 搜尋最近兩週內發行的專輯
* q=genre:Jazz 搜尋特定風格的藝術家
* q=year:1990-2000 搜尋 1990 至 2000 年間的專輯
* q=remaster%20track:Doxy%20artist:Miles%20Davis 找到 Miles Davis 的一首名為 Doxy 的歌曲（以重製版為主）

過濾條件適用範圍
不同條件只能在特定類型的搜尋中使用，例如：

* tag:new 和 tag:hipster 只適用於專輯。
* isrc 和 track 只適用於歌曲。
* genre 適用於藝術家和歌曲。
* 其他條件如 year 和 artist 可跨多類型使用。
* 查詢格式範例
q={keyword}%20filter:filter_value

關鍵字與過濾條件間使用空格 %20 連接。
多條件過濾時，條件之間用空格隔開。
